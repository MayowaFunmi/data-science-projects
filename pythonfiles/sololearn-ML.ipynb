# Build a logistic regression model with sklearn

```python
# import the Logistic Regression model
from sklearn.linear_model import LogisticRegression
```

```python
# instantiate the class
model = LogisticRegression()
```

```python
# an assumption test
a = df[['Fare', 'Age']].values # features
b = df['Survived'].values # target
model.fit(a, b)
```

```python
print(model.coef_, model.intercept_)
```

```python
# this means the equation is 
# 0 = 0.01615949x + -0.01549065y + -0.51037152
```

```python
# make predictions with models

x = df[['Pclass', 'male', 'Age', 'Siblings/Spouses', 'Parents/Children', 'Fare']].values
y = df['Survived'].values
model = LogisticRegression()
model.fit(x, y)
```

```python
model.predict(x)
```

```python
# predict if the first passenger survived or not.
# 1 means survived, 0 means not survived
model.predict([[3, True, 22.0, 1, 0, 7.25]]) # not survived
```

```python
# predict second passenger
model.predict([[1, False, 38.0, 1, 0, 71.2833]]) # survived
```

```python
# predict first 5
model.predict(x[:5])
```

```python
model.predict([[2, False, 45.0, 1, 1, 8.435]])
```

# Scoring the model # accuracy

```python
# create an array that has the predicted y values
y_pred = model.predict(x)
```

```python
y_pred # model's prediction
```

```python
# correct values
y
```

```python
# compare model's value and the correct values
y == y_pred
```

```python
# get the number of correct predictions
print((y == y_pred).sum())
```

```python
# get the percentage of accuracy
# total correct predictions x total number of passengers
print((y == y_pred).sum() / y.shape[0])
```

```python
# another method
print(model.score(x, y))
```
