# DataFrames

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
mydata = pd.DataFrame(data)
mydata
```

```python
# locate row = return one or more specified row(s)
mydata.loc[1]
```

```python
mydata.loc[[1, 2]]
```

```python
mydata = pd.DataFrame(data, index = ["day1", "day2", "day3"])
mydata
```

```python
# locate named index
mydata.loc['day3']
```
