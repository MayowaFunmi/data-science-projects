# creating a scatter plot using a data from csv file:

```python
# Bar chart or bar plot
```

```python
# vertical bars
subject = ['Mathematics', 'Physics', 'Chemistry', 'Biology']
marks = [90, 70, 75, 85]
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.bar(subject, marks) # subjects=horizontal, marks=vertical
plt.show()
```

```python
# horizontal bars
subject = ['Mathematics', 'Physics', 'Chemistry', 'Biology']
marks = [90, 70, 75, 85]
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.barh(subject, marks) # marks=horizontal, subjects=vertical
plt.show()
```

# Comparing two bar charts

```python
plt.bar([1, 3, 5, 7, 9], [5, 2, 7, 8, 3], label='Example One')
plt.bar([2, 4, 6, 8, 10], [8, 6, 2, 5, 7], label='Example Two', color='green')
plt.legend()
plt.xlabel('Bar Number')
plt.ylabel('Bar Height')
plt.title('Comparison Between Two Bar Graphs')
plt.show()
```

# Contour plots

```python
import numpy as np
# generating 100 numbers from -3.0 to 3.0
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp) # add a colorbar to the plot
plt.show()
```

# 3D Line Plot

```python
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
ax.plot3D(x, y, z, 'gray')
ax.set_title('3D line plot')
plt.show()
```

```python
ax = plt.axes(projection='3d')
z = np.linspace(1, 100, 100)
x = z * np.cos(200 * z)
y = z * np.sin(200 * z)
ax.plot3D(x, y, z, 'red')
ax.set_title('3D line plot')
plt.show()
```

```python
ax = plt.axes(projection='3d')
z = [5, 10, 15, 20]
x = [1, 2, 3, 4]
y = [100, 200, 300, 400]
ax.plot3D(x, y, z, 'red')
plt.show()
```
